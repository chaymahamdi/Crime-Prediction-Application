# Data cleaning

```python
# Import the data
df = pd.read_csv('../NY_crimes/NYPD_Complaint_Data_Historic.csv')
```

```python
df.info()
df.head()
```

```python
df.describe().T
```

```python
# Remove unnecessary columns
df.drop(columns=['RPT_DT', 'PD_CD', 'PD_DESC', 'KY_CD', 'JURISDICTION_CODE', 'JURIS_DESC', 'LOC_OF_OCCUR_DESC'], inplace=True)
```

```python
# Ratio of NaN values to total values for each column
for column in df.columns:
    nan_count = df[column].isna().sum()
    total_count = len(df[column])
    ratio = nan_count / total_count if total_count != 0 else 0  # Avoid division by zero
    print(f'{column}: {nan_count}/{total_count} NaN ratio: {ratio:.2%}')
```

```python
# Remove columns with significant NaN values and redundant information
df.drop(columns=['TRANSIT_DISTRICT', 'HOUSING_PSA'], inplace=True)
```

```python
# Converting notnull to 1 and isna to 0 for specific columns
df['IN_PARK'] = df['PARKS_NM'].notnull().astype(int)
df['IN_PUBLIC_HOUSING'] = df['HADEVELOPT'].notnull().astype(int)
df['IN_STATION'] = df['STATION_NAME'].notnull().astype(int)

# Renaming columns
df = df.rename({'PARKS_NM': 'IN_PARK', 'HADEVELOPT': 'IN_PUBLIC_HOUSING', 'STATION_NAME': 'IN_STATION'}, axis=1)
```

```python
# Dropping unnecessary coordinate columns
df.drop(columns=['Y_COORD_CD', 'Lat_Lon', 'X_COORD_CD'], inplace=True)
```

```python
# Dropping rows without coordinates
df = df[df['Latitude'].notnull()]
df = df[df['Longitude'].notnull()]
```

```python
# Check if two columns have matching NaN values
if all(df['CMPLNT_FR_DT'].notnull() == df['CMPLNT_TO_DT'].notnull()):
    df.drop(columns='CMPLNT_TO_DT', inplace=True)
```

```python
# Convert date columns to YYYY-MM-DD format
df['CMPLNT_FR_DT'] = pd.to_datetime(df['CMPLNT_FR_DT'], errors='coerce').dt.strftime('%Y-%m-%d')
df['CMPLNT_FR_TM'] = pd.to_datetime(df['CMPLNT_FR_TM'], errors='coerce').dt.strftime('%Y-%m-%d')
```

```python
# Delete rows without event start date or time
df = df[df['CMPLNT_FR_DT'].notnull() & df['CMPLNT_FR_TM'].notnull()]
```

```python
# Removing columns with redundant information
df.drop(columns=['CMPLNT_TO_TM', 'CMPLNT_TO_DT'], inplace=True)
```

```python
# Splitting date to year, month, day, and hour columns
df['year'] = pd.to_datetime(df['CMPLNT_FR_DT']).dt.year
df['month'] = pd.to_datetime(df['CMPLNT_FR_DT']).dt.month
df['day'] = pd.to_datetime(df['CMPLNT_FR_DT']).dt.day
df['hour'] = pd.to_datetime(df['CMPLNT_FR_TM']).dt.hour

# Adding a weekday column
df['weekday'] = pd.to_datetime(df['CMPLNT_FR_DT']).dt.day_name()

# Dropping original date and time columns
df.drop(columns=['CMPLNT_FR_DT', 'CMPLNT_FR_TM'], inplace=True)
```

```python
df.head()
```

```python
# Cleaning suspect and victim data by filling missing values
df['SUSP_RACE'].fillna('UNKNOWN', inplace=True)
df['VIC_RACE'].fillna('UNKNOWN', inplace=True)

df['SUSP_SEX'].fillna('U', inplace=True)
df['VIC_SEX'].fillna('U', inplace=True)
```

```python
valid_age = ['UNKNOWN', '25-44', '<18', '45-64', '65+', '18-24']

# Handling invalid age data
df.loc[~df['SUSP_AGE_GROUP'].isin(valid_age), 'SUSP_AGE_GROUP'] = 'UNKNOWN'
df.loc[~df['VIC_AGE_GROUP'].isin(valid_age), 'VIC_AGE_GROUP'] = 'UNKNOWN'
```

```python
# Removing redundant data and filling missing values
df.drop(columns='PATROL_BORO', inplace=True)
df['BORO_NM'].fillna('UNKNOWN', inplace=True)
```

```python
df.head()
```

```python
# Checking the remaining row count
len(df)
```

```python
# Encoding crime severity levels
df['LAW_CAT_CD'] = df['LAW_CAT_CD'].replace(['felony', 'misdemeanor', 'violation'], [2, 1, 0])
```

```python
# Displaying column names in the DataFrame
df.columns
```

```python
# Renaming column 'LAW_CAT_CD' to 'CRIME_CLASS'
df = df.rename({'LAW_CAT_CD': 'CRIME_CLASS'}, axis=1)
```

```python
# Counting occurrences of different crime classes
df['CRIME_CLASS'].value_counts()
```

```python
# Dropping duplicate columns if present
df = df.loc[:,~df.columns.duplicated()]

# Reordering columns in the DataFrame
column_titles = ['CMPLNT_NUM','year','month', 'day','weekday','hour','Latitude', 'Longitude', 'CRM_ATPT_CPTD_CD',
                 'OFNS_DESC', 'ADDR_PCT_CD', 'CRIME_CLASS', 'BORO_NM', 'PREM_TYP_DESC',
                 'IN_PARK', 'IN_PUBLIC_HOUSING', 'IN_STATION',
                 'SUSP_AGE_GROUP', 'SUSP_RACE', 'SUSP_SEX',
                 'VIC_AGE_GROUP', 'VIC_RACE', 'VIC_SEX']

df = df.reindex(columns=column_titles)
```

```python
df.head()
```

```python
# Filtering data for years greater than 2005
df = df[df['year'] > 2005]

# Creating separate DataFrames for 2019 and the training data (excluding 2019)
df_2019 = df[df['year'] == 2019]
df_train = df[df['year'] != 2019]
```

```python
# Save processed data to CSV files
df.to_csv('ny_processed_all.csv', index=False)
df_train.to_csv('ny_processed.csv', index=False)
df_2019.to_csv('ny_processed_2019.csv', index=False)
```
